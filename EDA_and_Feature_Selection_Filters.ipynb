#Feature Engineering

#1. ScratchDirectionEncoded
The ScratchDirectionEncoded feature captures the dominant orientation of the defect pattern within each wafer. It is computed by analyzing the spatial extent of bad dies (excluding good ones) and classifying the overall direction into one of six discrete types: 'horizontal', 'vertical', 'diagonal', 'mixed', 'point', or 'none'. This categorical signal is then numerically encoded and assigned to all dies in the wafer. By providing a coarse geometric context of the scratch shape, this feature helps the model differentiate between systematic linear patterns (e.g., horizontal scratch) and irregular or isolated noise, improving the accuracy of defect detection and classification.



```python
def classify_direction(dx, dy, threshold=3):
    if dx == 0 and dy == 0:
        return 'point'
    angle = math.degrees(math.atan2(dy, dx))
    angle = abs(angle) % 180  # remove negative and wrap

    if abs(angle - 0) < threshold or abs(angle - 180) < threshold:
        return 'horizontal'
    elif abs(angle - 90) < threshold:
        return 'vertical'
    elif abs(angle - 45) < threshold or abs(angle - 135) < threshold:
        return 'diagonal'
    else:
        return 'mixed'

# Assign estimated defect direction based only on bad dies
for name, df in wafer_dict.items():
    defects = df[df['IsGoodDie'] == 0]
    if len(defects) >= 2:
        dx = defects['DieX'].max() - defects['DieX'].min()
        dy = defects['DieY'].max() - defects['DieY'].min()
        direction = classify_direction(dx, dy)
    elif len(defects) == 1:
        direction = 'point'
    else:
        direction = 'none'

    # Assign the same direction to all dies in the wafer
    wafer_dict[name]['DefectDirection'] = direction

# ✅ Verification
unique_values = set()
for name, df in wafer_dict.items():
    if 'DefectDirection' not in df.columns:
        print(f"❌ Missing DefectDirection in wafer: {name}")
    else:
        unique_values.update(df['DefectDirection'].unique())

print("\n✅ All wafers now have 'DefectDirection' column.")
print(f"🧭 Found direction types: {unique_values}")
```

```python
direction_map = {
    'none': 0,
    'point': 1,
    'horizontal': 2,
    'vertical': 3,
    'diagonal': 4,
    'mixed': 5
}

# Apply the mapping to each wafer
for name, df in wafer_dict.items():
    direction = df['DefectDirection'].iloc[0]
    encoded_value = direction_map.get(direction, -1)
    wafer_dict[name]['DefectDirectionEncoded'] = encoded_value

# ✅ Verification
print("✅ Added 'DefectDirectionEncoded' column to all wafers.")
print(f"🔢 Direction mapping used: {direction_map}")
```

#2.Geometric and Wafer-Level Attributes
In this phase, we engineered fundamental geometric features for each die, including normalized coordinates (NormDieX, NormDieY), radial distance from the wafer center (DistanceFromCenter), angular position (AngleFromCenter), and edge proximity (IsEdge). Additionally, we computed global wafer-level attributes such as wafer size, yield rate, and defect direction encoding. These features jointly capture the local spatial positioning of each die and the broader structural context of its wafer, enabling the model to better identify elongated defect patterns characteristic of scratches.



```python

def build_geometric_features(df_wafer):
    df = df_wafer.copy()

    # Normalize coordinates
    dx_range = df['DieX'].max() - df['DieX'].min()
    dy_range = df['DieY'].max() - df['DieY'].min()
    df['NormDieX'] = (df['DieX'] - df['DieX'].min()) / dx_range
    df['NormDieY'] = (df['DieY'] - df['DieY'].min()) / dy_range

    # Distance & angle from center
    df['DistanceFromCenter'] = np.sqrt((df['NormDieX'] - 0.5)**2 + (df['NormDieY'] - 0.5)**2)
    df['AngleFromCenter'] = np.arctan2(df['NormDieY'] - 0.5, df['NormDieX'] - 0.5)

    # Dynamic margin for edges
    dynamic_margin = 1 / max(dx_range, dy_range) * 5
    df['IsEdge'] = (
        (df['NormDieX'] < dynamic_margin) | (df['NormDieX'] > 1 - dynamic_margin) |
        (df['NormDieY'] < dynamic_margin) | (df['NormDieY'] > 1 - dynamic_margin)
    ).astype(int)

    # IsCenterLine feature (close to center axis)
    df['IsCenterLine'] = np.logical_or(
    np.abs(df['NormDieX'] - 0.5) < 0.05,
    np.abs(df['NormDieY'] - 0.5) < 0.05
).astype(int)

    # Quadrant and radial band
    df['Quadrant'] = (
        (df['NormDieX'] > 0.5).astype(int) +
        2 * (df['NormDieY'] > 0.5).astype(int)
    )
    df['RadialBand'] = pd.cut(df['DistanceFromCenter'], bins=5, labels=False)

    # Wafer-level features
    df['WaferSize'] = len(df)
    df['WaferYield'] = df['IsGoodDie'].mean()
    if 'DefectDirectionEncoded' in df.columns:
        df['DefectDirectionEncoded'] = df['DefectDirectionEncoded'].iloc[0]
    else:
        df['DefectDirectionEncoded'] = -1  # fallback

    # One-hot encode Quadrant and RadialBand
    df = pd.get_dummies(df, columns=['Quadrant', 'RadialBand'], prefix=['Quad', 'Band'])

    return df


# Apply to all wafers
for name in wafer_dict:
    wafer_dict[name] = build_geometric_features(wafer_dict[name])

print("✅ All wafers updated in wafer_dict with enhanced geometric features.")
```

#3.Neighborhood Features

In this phase, we engineered advanced neighborhood features to capture the local spatial context of each die. For both 3×3 and 5×5 neighborhoods, we calculated the number and ratio of defective (BadNeighbors) and good dies (GoodNeighbors), as well as derived indicators such as whether all surrounding dies are bad (AllBad), and the weighted sum of nearby defects based on proximity (BadWeightedSum). To capture scratch-like patterns, we also computed the longest streaks of defective dies in horizontal and vertical directions (BadStreakH, BadStreakV), and detected diagonal defect structures (DiagonalBadPair). These features enrich the model’s ability to identify elongated or clustered defects consistent with scratch behavior, without relying on the scratch label itself.



```python
def build_neighborhood_features(df_wafer, radii=[1, 2]):
    """
    Compute neighborhood features for given wafer DataFrame.
    Includes advanced spatial context features.
    """
    df = df_wafer.copy()
    die_map = {
        (row.DieX, row.DieY): row.IsGoodDie
        for row in df.itertuples()
    }

    for r in radii:
        max_neighbors = (2*r + 1)**2 - 1  # excluding center

        # Containers
        bad_list = []
        good_list = []
        ratio_list = []
        all_bad_list = []
        weighted_bad_list = []
        weighted_good_list = []
        streak_h_list = []
        streak_v_list = []
        diag_pair_list = []
        max_streak_dir = []  # direction with longest streak

        for row in df.itertuples():
            x, y = row.DieX, row.DieY
            bad = 0
            good = 0
            weighted_bad = 0
            weighted_good = 0
            diag_bad_pairs = 0
            max_h_streak = 0
            max_v_streak = 0

            for dx in range(-r, r + 1):
                h_streak = 0
                v_streak = 0
                for dy in range(-r, r + 1):
                    if dx == 0 and dy == 0:
                        continue
                    neighbor = (x + dx, y + dy)
                    if neighbor in die_map:
                        is_good = die_map[neighbor]
                        distance = np.sqrt(dx**2 + dy**2)
                        weight = 1 / distance if distance != 0 else 0

                        if is_good == 0:
                            bad += 1
                            weighted_bad += weight

                            if dy == 0:  # horizontal streak
                                h_streak += 1
                                max_h_streak = max(max_h_streak, h_streak)
                            if dx == 0:  # vertical streak
                                v_streak += 1
                                max_v_streak = max(max_v_streak, v_streak)
                        else:
                            good += 1
                            weighted_good += weight
                # Reset after dx sweep
                h_streak = 0
                v_streak = 0

            # Diagonal bad pair detection
            diag_coords = [((x-1, y-1), (x+1, y+1)), ((x-1, y+1), (x+1, y-1))]
            has_diag = 0
            for a, b in diag_coords:
                if die_map.get(a, 1) == 0 and die_map.get(b, 1) == 0:
                    has_diag = 1
                    break

            total = bad + good
            bad_ratio = bad / total if total > 0 else 0
            all_bad = int(bad > 0 and good == 0)

            # Which direction has strongest streak?
            direction = 'horizontal' if max_h_streak > max_v_streak else 'vertical'
            if max_h_streak == 0 and max_v_streak == 0 and has_diag:
                direction = 'diagonal'
            if max_h_streak == 0 and max_v_streak == 0 and not has_diag:
                direction = 'none'

            # Append results
            bad_list.append(bad)
            good_list.append(good)
            ratio_list.append(bad_ratio)
            all_bad_list.append(all_bad)
            weighted_bad_list.append(weighted_bad)
            weighted_good_list.append(weighted_good)
            streak_h_list.append(max_h_streak)
            streak_v_list.append(max_v_streak)
            diag_pair_list.append(has_diag)
            max_streak_dir.append(direction)

        suffix = f"{2*r+1}x{2*r+1}"
        df[f'BadNeighbors_{suffix}'] = bad_list
        df[f'GoodNeighbors_{suffix}'] = good_list
        df[f'BadRatio_{suffix}'] = ratio_list
        df[f'AllBad_{suffix}'] = all_bad_list
        df[f'BadWeightedSum_{suffix}'] = weighted_bad_list
        df[f'GoodWeightedSum_{suffix}'] = weighted_good_list
        df[f'BadNeighborsNorm_{suffix}'] = np.array(bad_list) / max_neighbors
        df[f'BadStreakH_{suffix}'] = streak_h_list
        df[f'BadStreakV_{suffix}'] = streak_v_list
        df[f'DiagonalBadPair_{suffix}'] = diag_pair_list

        # Encode direction numerically
        dir_map = {'none': 0, 'horizontal': 1, 'vertical': 2, 'diagonal': 3}
        df[f'MaxStreakDirection_{suffix}'] = [dir_map[d] for d in max_streak_dir]

    return df




for name in wafer_dict:
    wafer_dict[name] = build_neighborhood_features(wafer_dict[name], radii=[1, 2])


print("✅ Neighborhood features's added to all wafers.")
```

#4.Streak-based Features

In this phase, we introduced streak-based features designed to detect linear sequences of defective dies, which are indicative of scratches. For each die, we computed the number of consecutive bad neighbors in four directions: horizontal, vertical, and both diagonals. We then derived the MaxBadStreak feature, representing the longest such sequence per die, and a normalized version relative to the wafer size (MaxBadStreakNorm). These features enable the model to recognize elongated and directional defect patterns that may not be fully continuous but strongly resemble physical scratches.



```python
def build_streak_features(df_wafer):
    df = df_wafer.copy()

    bad_set = set(
        (row.DieX, row.DieY)
        for row in df.itertuples()
        if not row.IsGoodDie
    )

    def count_streak(x, y, dx, dy):
        count = 0
        i = 1
        while (x + dx * i, y + dy * i) in bad_set:
            count += 1
            i += 1
        i = 1
        while (x - dx * i, y - dy * i) in bad_set:
            count += 1
            i += 1
        return count

    horiz, vert, diag_pos, diag_neg = [], [], [], []

    for row in df.itertuples():
        x, y = row.DieX, row.DieY
        horiz.append(count_streak(x, y, 1, 0))       # Horizontal
        vert.append(count_streak(x, y, 0, 1))        # Vertical
        diag_pos.append(count_streak(x, y, 1, 1))    # Diagonal \
        diag_neg.append(count_streak(x, y, 1, -1))   # Diagonal /

    df['HorizontalBadStreak'] = horiz
    df['VerticalBadStreak'] = vert
    df['DiagonalBadStreak_Pos'] = diag_pos
    df['DiagonalBadStreak_Neg'] = diag_neg

    # Max streak & normalized
    df['MaxBadStreak'] = df[[
        'HorizontalBadStreak',
        'VerticalBadStreak',
        'DiagonalBadStreak_Pos',
        'DiagonalBadStreak_Neg'
    ]].max(axis=1)

    wafer_size = len(df)
    df['MaxBadStreakNorm'] = df['MaxBadStreak'] / wafer_size if wafer_size > 0 else 0

    # Direction (optional)
    df['DominantStreakDirection'] = df[[
        'HorizontalBadStreak',
        'VerticalBadStreak',
        'DiagonalBadStreak_Pos',
        'DiagonalBadStreak_Neg'
    ]].idxmax(axis=1).str.replace('BadStreak_', '').str.lower()

    direction_map = {
        'horizontal': 1,
        'vertical': 2,
        'diagonalbadstreak_pos': 3,
        'diagonalbadstreak_neg': 4
    }
    df['DominantStreakDirEncoded'] = df['DominantStreakDirection'].map(direction_map).fillna(0).astype(int)

    return df


# Apply to all wafers
for name in wafer_dict:
    wafer_dict[name] = build_streak_features(wafer_dict[name])

print("✅ Streak features (basic + max + normalized) added to all wafers.")
```

#5.AngleFromCenter
We extended the add_angle_from_center function to include additional polar geometric features. In addition to computing the angle of each die from the wafer center, we added trigonometric encodings (sin, cos), a discrete sector index (AngleSector_8), and a combined polar region identifier (PolarTile) by joining the radial and angular bins. These features provide the model with rich spatial context, enabling detection of direction- or region-specific scratch patterns.



```python
import numpy as np
def add_angle_from_center(df_wafer):
    df = df_wafer.copy()

    x_centered = df['NormDieX'] - 0.5
    y_centered = df['NormDieY'] - 0.5

    angle = np.arctan2(y_centered, x_centered)
    df['AngleFromCenter'] = angle
    df['AngleFromCenter_sin'] = np.sin(angle)
    df['AngleFromCenter_cos'] = np.cos(angle)

    df['AngleSector_8'] = ((8 * (angle + np.pi) / (2 * np.pi)).astype(int))  # 0–7

    if 'RadialBand' not in df.columns:
        df['DistanceFromCenter'] = np.sqrt(x_centered**2 + y_centered**2)
        df['RadialBand'] = pd.cut(df['DistanceFromCenter'], bins=5, labels=False)

    polar_tile = df['RadialBand'].astype(str) + "_" + df['AngleSector_8'].astype(str)
    df['PolarTileEncoded'] = polar_tile.astype('category').cat.codes

    return df



# Apply to all wafers
for name in wafer_dict:
    wafer_dict[name] = add_angle_from_center(wafer_dict[name])

print("✅ Added feature: 'AngleFromCenter'")
```

#6. IsOnEdge
We extended the edge feature module to capture multiple levels of edge-related context. In addition to detecting whether a die lies directly on the wafer's boundary (IsOnEdge), we added softer margin detection (IsNearEdge), corner indicators (IsCorner), and a graded edge depth measure (EdgeBandIndex) based on distance from the wafer center. These features provide fine-grained spatial awareness, helping the model capture mechanical or handling-related edge effects that are commonly associated with scratches.



```python
def add_edge_feature(df_wafer):
    df = df_wafer.copy()

    min_x, max_x = df['DieX'].min(), df['DieX'].max()
    min_y, max_y = df['DieY'].min(), df['DieY'].max()

    df['IsOnEdge'] = (
        (df['DieX'] == min_x) |
        (df['DieX'] == max_x) |
        (df['DieY'] == min_y) |
        (df['DieY'] == max_y)
    ).astype(int)

    margin = 0.05
    df['IsNearEdge'] = (
        (df['NormDieX'] < margin) | (df['NormDieX'] > 1 - margin) |
        (df['NormDieY'] < margin) | (df['NormDieY'] > 1 - margin)
    ).astype(int)

    df['IsCorner'] = (
        ((df['NormDieX'] < margin) & (df['NormDieY'] < margin)) |
        ((df['NormDieX'] < margin) & (df['NormDieY'] > 1 - margin)) |
        ((df['NormDieX'] > 1 - margin) & (df['NormDieY'] < margin)) |
        ((df['NormDieX'] > 1 - margin) & (df['NormDieY'] > 1 - margin))
    ).astype(int)

    if 'DistanceFromCenter' in df.columns:
        df['EdgeBandIndex'] = pd.cut(
            df['DistanceFromCenter'],
            bins=[0.85, 0.9, 0.95, 0.98, 1.0],
            labels=[1, 2, 3, 4],
            right=True
        ).astype('Int64')
    else:
        df['EdgeBandIndex'] = pd.NA

    return df

# Apply to all wafers
for name in wafer_dict:
    wafer_dict[name] = add_edge_feature(wafer_dict[name])

print("✅ Added edge-related features: IsOnEdge, IsNearEdge, IsCorner, EdgeBandIndex")
```

#.7 Asymmetry
We added asymmetry-based features to capture the spatial deviation of each die from the wafer’s diagonal axis. Asymmetry reflects the absolute difference between the die’s X and Y positions, while AsymmetryNorm and DiagonalDistance provide a normalized measure of deviation from the center diagonal. These features help the model detect off-axis patterns, particularly relevant for identifying diagonal or irregular scratch formations.










```python
def add_asymmetry_feature(df_wafer):
    df = df_wafer.copy()

    # Absolute difference between X and Y positions
    df['Asymmetry'] = np.abs(df['DieX'] - df['DieY'])

    # Normalized asymmetry (requires normalized coordinates)
    if 'NormDieX' in df.columns and 'NormDieY' in df.columns:
        df['AsymmetryNorm'] = np.abs(df['NormDieX'] - df['NormDieY'])
        df['DiagonalDistance'] = df['AsymmetryNorm']  # optional alias

    else:
        df['AsymmetryNorm'] = pd.NA
        df['DiagonalDistance'] = pd.NA

    return df


# Apply to all wafers
for name in wafer_dict:
    wafer_dict[name] = add_asymmetry_feature(wafer_dict[name])

print("✅ Added asymmetry features: Asymmetry, AsymmetryNorm, DiagonalDistance")
```

#8. ManhattanDistanceFromCenter
We introduced ManhattanDistanceFromCenter to reflect grid-based distance from the wafer center, which better aligns with the rectangular layout of dies. We also computed a discrete ManhattanRank to group dies into concentric rectangular layers, and a DistanceRatio comparing Euclidean and Manhattan distances, offering insights into geometric distortion or off-axis positioning.



```python
def add_manhattan_distance(df_wafer):
    df = df_wafer.copy()

    # Manhattan distance from center
    dx = np.abs(df['NormDieX'] - 0.5)
    dy = np.abs(df['NormDieY'] - 0.5)
    manhattan = dx + dy
    df['ManhattanDistanceFromCenter'] = manhattan

    # Manhattan-based rank (useful for identifying rectangular "rings")
    df['ManhattanRank'] = pd.qcut(manhattan, q=5, labels=False)

    # Distance ratio: Euclidean / Manhattan (requires DistanceFromCenter)
    if 'DistanceFromCenter' in df.columns:
        df['DistanceRatio_Euclid_vs_Manhattan'] = df['DistanceFromCenter'] / (manhattan + 1e-6)
    else:
        df['DistanceRatio_Euclid_vs_Manhattan'] = pd.NA

    return df


# Apply to all wafers
for name in wafer_dict:
    wafer_dict[name] = add_manhattan_distance(wafer_dict[name])

print("✅ Added feature: 'ManhattanDistanceFromCenter'")
```

#10.DistanceToNearestOppositeLabel
We added DistanceToNearestOppositeLabel to measure the Euclidean distance from each die to the closest die with the opposite IsGoodDie label. This feature captures the spatial transitions between good and bad die clusters, helping the model detect edge zones and potential scratch boundaries. It is especially useful for identifying good dies surrounded by nearby defects, which are candidates for inking.



```python

def add_distance_to_opposite_label(df_wafer):
    df = df_wafer.copy()

    good_coords = df[df['IsGoodDie'] == True][['DieX', 'DieY']].values
    bad_coords = df[df['IsGoodDie'] == False][['DieX', 'DieY']].values

    # Build KD-Trees for fast nearest neighbor search
    if len(good_coords) > 0:
        good_tree = cKDTree(good_coords)
    if len(bad_coords) > 0:
        bad_tree = cKDTree(bad_coords)

    distances = []

    for row in df.itertuples():
        coord = np.array([row.DieX, row.DieY])

        if row.IsGoodDie == True:
            if len(bad_coords) > 0:
                dist, _ = bad_tree.query(coord, k=1)
                distances.append(dist)
            else:
                distances.append(np.nan)
        else:
            if len(good_coords) > 0:
                dist, _ = good_tree.query(coord, k=1)
                distances.append(dist)
            else:
                distances.append(np.nan)

    df['DistanceToNearestOppositeLabel'] = distances
    return df

# Apply to all wafers
for name in wafer_dict:
    wafer_dict[name] = add_distance_to_opposite_label(wafer_dict[name])

print("✅ Added feature: 'DistanceToNearestOppositeLabel'")
```

#SecondNearestBadDistance and DistanceToBadClusterCenter
We introduced two legal spatial proximity features to replace scratch-based ones. SecondNearestBadDistance measures the distance to the second-closest defective die (IsGoodDie == False), while DistanceToBadClusterCenter captures the distance from each die to the centroid of all bad dies in the wafer. These features approximate defect clustering patterns while remaining valid in both train and test scenarios.





```python

def add_proximity_to_bad_clusters(df_wafer):
    df = df_wafer.copy()

    bad_coords = df[df['IsGoodDie'] == False][['DieX', 'DieY']].values
    all_coords = df[['DieX', 'DieY']].values

    if len(bad_coords) == 0:
        df['SecondNearestBadDistance'] = np.nan
        df['DistanceToBadClusterCenter'] = np.nan
        return df

    # Compute all distances between all dies and bad dies
    distances_matrix = cdist(all_coords, bad_coords)

    # Take second smallest distance for each row
    sorted_dists = np.sort(distances_matrix, axis=1)
    second_dists = sorted_dists[:, 1] if bad_coords.shape[0] > 1 else np.full(len(all_coords), np.nan)
    df['SecondNearestBadDistance'] = second_dists

    # Compute bad cluster center and distance to it
    bad_center = np.mean(bad_coords, axis=0)
    center_dists = np.linalg.norm(all_coords - bad_center, axis=1)
    df['DistanceToBadClusterCenter'] = center_dists

    return df


for name in wafer_dict:
    wafer_dict[name] = add_proximity_to_bad_clusters(wafer_dict[name])

print("✅ Added features: SecondNearestBadDistance, DistanceToBadClusterCenter ")

```

#11.WeightedBadNeighborScore
The WeightedBadNeighborScore is now efficiently computed using pairwise distance matrices. It estimates local risk for each die based on nearby defective dies, assigning higher weight to closer neighbors. This fast, vectorized version scales better and maintains high spatial resolution.




```python

def add_weighted_bad_neighbor_score(df_wafer, radius=3):
    df = df_wafer.copy()

    all_coords = df[['DieX', 'DieY']].values
    bad_coords = df[df['IsGoodDie'] == False][['DieX', 'DieY']].values

    if len(bad_coords) == 0:
        df['WeightedBadNeighborScore'] = 0
        return df

    # Compute pairwise distances between all dies and bad dies
    dist_matrix = cdist(all_coords, bad_coords)

    # Mask: keep distances within radius and >0
    with np.errstate(divide='ignore'):
        weights = np.where((dist_matrix > 0) & (dist_matrix <= radius), 1 / dist_matrix, 0)

    # Sum the weighted contributions
    weighted_scores = np.sum(weights, axis=1)
    df['WeightedBadNeighborScore'] = weighted_scores

    return df


# Apply to all wafers
for name in wafer_dict:
    wafer_dict[name] = add_weighted_bad_neighbor_score(wafer_dict[name])

print("✅ Added feature: 'WeightedBadNeighborScore'")
```

#12.add_multi_radius_neighbor_counts
This function computes spatial neighbor counts within multiple radii, capturing both the number of nearby defective dies (BadNeighborRadiusCount_r{r}) and the number of good dies (GoodNeighborRadiusCount_r{r}). It also computes their ratio (BadGoodRatio_r{r}), providing the model with valuable insight into the local balance between good and bad dies. These features enable the detection of high-risk zones where healthy dies are surrounded by defects, supporting precise scratch detection.






```python

def add_multi_radius_neighbor_counts(df_wafer, radii=[1, 2, 3]):
    df = df_wafer.copy()

    all_coords = df[['DieX', 'DieY']].values
    bad_coords = df[df['IsGoodDie'] == False][['DieX', 'DieY']].values
    good_coords = df[df['IsGoodDie'] == True][['DieX', 'DieY']].values

    # Compute distance matrices once
    bad_dists = cdist(all_coords, bad_coords) if len(bad_coords) > 0 else None
    good_dists = cdist(all_coords, good_coords) if len(good_coords) > 0 else None

    for r in radii:
        # Bad neighbor count
        if bad_dists is not None:
            bad_count = ((bad_dists > 0) & (bad_dists <= r)).sum(axis=1)
        else:
            bad_count = np.zeros(len(all_coords), dtype=int)
        df[f'BadNeighborRadiusCount_r{r}'] = bad_count

        # Good neighbor count
        if good_dists is not None:
            good_count = ((good_dists > 0) & (good_dists <= r)).sum(axis=1)
        else:
            good_count = np.zeros(len(all_coords), dtype=int)
        df[f'GoodNeighborRadiusCount_r{r}'] = good_count

        # Ratio (safe division)
        df[f'BadGoodRatio_r{r}'] = bad_count / (good_count + 1)

    return df


# Apply to all wafers
for name in wafer_dict:
    wafer_dict[name] = add_multi_radius_neighbor_counts(wafer_dict[name], radii=[1, 2, 3])



print("✅ Added feature: 'BadNeighborRadiusCount_r3'")
```

#GoodNeighborRadiusCount
We introduced GoodNeighborRadiusCount_r{r} features to count the number of nearby good dies within a given radius. These features complement the bad neighbor counts, enabling the model to assess local quality density and identify outlier dies surrounded by a majority of either good or bad neighbors.



```python

def add_multi_radius_good_neighbor_counts(df_wafer, radii=[1, 2, 3]):
    df = df_wafer.copy()
    good_coords = df[df['IsGoodDie'] == True][['DieX', 'DieY']].values
    all_coords = df[['DieX', 'DieY']].values

    if len(good_coords) == 0:
        for r in radii:
            df[f'GoodNeighborRadiusCount_r{r}'] = 0
        return df

    dist_matrix = cdist(all_coords, good_coords)

    for r in radii:
        count_matrix = ((dist_matrix > 0) & (dist_matrix <= r)).sum(axis=1)
        df[f'GoodNeighborRadiusCount_r{r}'] = count_matrix

    return df
for name in wafer_dict:
    wafer_dict[name] = add_multi_radius_good_neighbor_counts(wafer_dict[name], radii=[1, 2, 3])
```

#13.LocalDefectDensity_KDE
This upgraded version of the KDE-based feature function includes a ratio metric LocalDefectToGoodDensityRatio, which compares smoothed defect density to good die density around each point. This contrastive feature enables the model to better discriminate between locally "safe" and "risky" zones, particularly in ambiguous edge or transition areas.




```python

def add_kde_density_full(df_wafer, normalize=True, bw_method='silverman'):
    df = df_wafer.copy()
    query_coords = df[['DieX', 'DieY']].values

    # === Bad Dies KDE ===
    bad_dies = df[df['IsGoodDie'] == False][['DieX', 'DieY']].values
    if len(bad_dies) >= 2:
        kde_bad = gaussian_kde(bad_dies.T, bw_method=bw_method)
        densities_bad = kde_bad(query_coords.T)
        densities_bad_norm = densities_bad / np.max(densities_bad) if normalize and np.max(densities_bad) > 0 else densities_bad
    else:
        densities_bad = np.zeros(len(df))
        densities_bad_norm = densities_bad

    df['LocalDefectDensity_KDE'] = densities_bad
    df['LocalDefectDensity_KDE_Norm'] = densities_bad_norm

    # === Good Dies KDE ===
    good_dies = df[df['IsGoodDie'] == True][['DieX', 'DieY']].values
    if len(good_dies) >= 2:
        kde_good = gaussian_kde(good_dies.T, bw_method=bw_method)
        densities_good = kde_good(query_coords.T)
        densities_good_norm = densities_good / np.max(densities_good) if normalize and np.max(densities_good) > 0 else densities_good
    else:
        densities_good = np.zeros(len(df))
        densities_good_norm = densities_good

    df['LocalGoodDensity_KDE'] = densities_good
    df['LocalGoodDensity_KDE_Norm'] = densities_good_norm
    # === Density Ratio: Bad / Good ===
    epsilon = 1e-6
    df['LocalDefectToGoodDensityRatio'] = densities_bad_norm / (densities_good_norm + epsilon)

    return df



# Apply to all wafers
for name in wafer_dict:
    wafer_dict[name] = add_kde_density_full(wafer_dict[name])



print("✅ Added feature: 'LocalDefectDensity_KDE'")
```

#14.Connected Component Features per Wafer
This comprehensive component feature extractor combines spatial, geometric, and topological analysis to characterize each defect cluster in detail. It includes compactness, linearity, circularity, bounding box metrics, center distance, and principal orientation, both in angle and categorical form. These features enable high-resolution modeling of scratch patterns, providing the model with rich structural context for accurate classification.





```python

def add_defect_component_features(df_wafer, min_cluster_size=4):
    df = df_wafer.copy()

    x_vals = df['DieX'].unique()
    y_vals = df['DieY'].unique()
    x_map = {x: i for i, x in enumerate(sorted(x_vals))}
    y_map = {y: i for i, y in enumerate(sorted(y_vals))}
    wafer_center = np.array([len(x_vals) / 2, len(y_vals) / 2])

    grid = np.zeros((len(y_vals), len(x_vals)), dtype=int)
    for row in df.itertuples():
        if not row.IsGoodDie:
            grid[y_map[row.DieY], x_map[row.DieX]] = 1

    labeled_grid, _ = label(grid, structure=np.ones((3, 3)))

    label_map = {}
    cluster_coords = {}
    for y in range(labeled_grid.shape[0]):
        for x in range(labeled_grid.shape[1]):
            lbl = labeled_grid[y, x]
            if lbl > 0:
                key = (x, y)
                label_map[key] = lbl
                cluster_coords.setdefault(lbl, []).append((x, y))

    stats = {}
    for lbl, coords in cluster_coords.items():
        coords_arr = np.array(coords)
        size = len(coords)
        center = coords_arr.mean(axis=0)
        xs, ys = coords_arr[:, 0], coords_arr[:, 1]
        width = xs.max() - xs.min() + 1
        height = ys.max() - ys.min() + 1
        aspect = max(width, height) / (min(width, height) + 1e-6)
        box_area = width * height
        compactness = size / (box_area + 1e-6)
        density = size / (box_area + 1e-6)
        center_distance = np.linalg.norm(center - wafer_center)

        # SVD for orientation and linearity
        coords_centered = coords_arr - center
        if coords_arr.shape[0] >= 2:
            U, S, Vh = np.linalg.svd(coords_centered)
            direction = Vh[0]
            orientation = np.arctan2(direction[1], direction[0])
            linearity = S[0] / (S[1] + 1e-6)
        else:
            orientation = 0.0
            linearity = 0.0

        # Orientation category
        angle_deg = np.degrees(orientation % np.pi)
        if angle_deg < 22.5 or angle_deg > 157.5:
            orientation_cat = 'horizontal'
        elif 67.5 < angle_deg <= 112.5:
            orientation_cat = 'vertical'
        elif 22.5 < angle_deg <= 67.5:
            orientation_cat = 'diagonal_pos'
        else:
            orientation_cat = 'diagonal_neg'

        # Perimeter estimate
        coords_set = set(map(tuple, coords))
        perimeter = 0
        for x, y in coords:
            neighbors = [(x+dx, y+dy) for dx, dy in [(-1,0),(1,0),(0,-1),(0,1)]]
            perimeter += sum((n not in coords_set) for n in neighbors)
        circularity = (4 * np.pi * size) / (perimeter**2 + 1e-6)

        stats[lbl] = {
            'size': size,
            'center': center,
            'width': width,
            'height': height,
            'aspect': aspect,
            'compactness': compactness,
            'density': density,
            'orientation': orientation,
            'linearity': linearity,
            'orientation_cat': orientation_cat,
            'perimeter': perimeter,
            'circularity': circularity,
            'center_distance': center_distance
        }

    # Assign to dataframe
    features = {
        'DefectComponentLabel': [],
        'DefectComponentSize': [],
        'IsInLargeDefectCluster': [],
        'DistanceToDefectClusterCenter': [],
        'DefectClusterAspectRatio': [],
        'DefectClusterOrientation': [],
        'DefectClusterOrientationCategory': [],
        'DefectClusterCompactness': [],
        'DefectClusterWidth': [],
        'DefectClusterHeight': [],
        'DefectClusterDensity': [],
        'DefectClusterLinearity': [],
        'DefectClusterPerimeter': [],
        'DefectClusterCircularity': [],
        'ClusterCenterDistanceFromWaferCenter': [],
    }

    for row in df.itertuples():
        cx = x_map[row.DieX]
        cy = y_map[row.DieY]
        key = (cx, cy)

        if key in label_map:
            lbl = label_map[key]
            st = stats[lbl]
            dist = np.linalg.norm(np.array([cx, cy]) - st['center'])
        else:
            lbl, st, dist = -1, {}, 0

        features['DefectComponentLabel'].append(lbl)
        features['DefectComponentSize'].append(st.get('size', 0))
        features['IsInLargeDefectCluster'].append(int(st.get('size', 0) >= min_cluster_size))
        features['DistanceToDefectClusterCenter'].append(dist)
        features['DefectClusterAspectRatio'].append(st.get('aspect', 0))
        features['DefectClusterOrientation'].append(st.get('orientation', 0))
        features['DefectClusterOrientationCategory'].append(st.get('orientation_cat', 'none'))
        features['DefectClusterCompactness'].append(st.get('compactness', 0))
        features['DefectClusterWidth'].append(st.get('width', 0))
        features['DefectClusterHeight'].append(st.get('height', 0))
        features['DefectClusterDensity'].append(st.get('density', 0))
        features['DefectClusterLinearity'].append(st.get('linearity', 0))
        features['DefectClusterPerimeter'].append(st.get('perimeter', 0))
        features['DefectClusterCircularity'].append(st.get('circularity', 0))
        features['ClusterCenterDistanceFromWaferCenter'].append(st.get('center_distance', 0))

    for col, values in features.items():
        df[col] = values

    return df



for name in wafer_dict:
    wafer_dict[name] = add_defect_component_features(wafer_dict[name])

print("✅ Added: 'DefectComponentLabel', 'DefectComponentSize', 'IsInLargeDefectCluster'")
```

#15.Graph-Based Features
We expanded the graph-based defect analysis with advanced structural metrics. These include local connectivity (Degree), clustering coefficient, articulation point detection, and node centrality measures (Betweenness, Closeness). Together, these features offer a detailed view of how each bad die is embedded in the global defect topology, supporting robust detection of scratch-like formations, bridge nodes, and core defect zones.



```python

def add_graph_based_features(df_wafer):
    df = df_wafer.copy()

    bad_df = df[df['IsGoodDie'] == False]
    bad_coords = set(zip(bad_df['DieX'], bad_df['DieY']))

    G = nx.Graph()

    for coord in bad_coords:
        G.add_node(coord)

    for x, y in bad_coords:
        for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            neighbor = (x + dx, y + dy)
            if neighbor in bad_coords:
                G.add_edge((x, y), neighbor)

    components = list(nx.connected_components(G))

    node_to_component_size = {}
    node_to_degree = {}
    node_to_cluster_coef = {}
    articulation_points = set(nx.articulation_points(G)) if nx.is_connected(G) else set()
    try:
        betweenness = nx.betweenness_centrality(G, normalized=True)
        closeness = nx.closeness_centrality(G)
    except:
        betweenness = {}
        closeness = {}

    for component in components:
        subgraph = G.subgraph(component)
        size = len(component)
        clustering = nx.clustering(subgraph)
        for node in component:
            node_to_component_size[node] = size
            node_to_degree[node] = G.degree[node]
            node_to_cluster_coef[node] = clustering.get(node, 0)

    # Assign features
    subgraph_size = []
    is_in_subgraph = []
    subgraph_degree = []
    subgraph_clustering = []
    is_articulation = []
    node_betweenness = []
    node_closeness = []

    for row in df.itertuples():
        coord = (row.DieX, row.DieY)
        subgraph_size.append(node_to_component_size.get(coord, 0))
        is_in_subgraph.append(int(coord in node_to_component_size))
        subgraph_degree.append(node_to_degree.get(coord, 0))
        subgraph_clustering.append(node_to_cluster_coef.get(coord, 0))
        is_articulation.append(int(coord in articulation_points))
        node_betweenness.append(betweenness.get(coord, 0))
        node_closeness.append(closeness.get(coord, 0))

    df['DefectSubgraphSize'] = subgraph_size
    df['IsInDefectSubgraph'] = is_in_subgraph
    df['DefectSubgraphDegree'] = subgraph_degree
    df['DefectSubgraphClustering'] = subgraph_clustering
    df['IsArticulationPoint'] = is_articulation
    df['DefectSubgraphBetweenness'] = node_betweenness
    df['DefectSubgraphCloseness'] = node_closeness

    return df

for name in wafer_dict:
    wafer_dict[name] = add_graph_based_features(wafer_dict[name])

print("✅ Added: 'DefectSubgraphSize', 'IsInDefectSubgraph', 'DefectSubgraphDegree'")
```

#16.Image-Pattern Features
This enriched image-pattern feature extractor utilizes regionprops from scikit-image to compute geometric shape descriptors for each connected defect region. These include classical metrics like eccentricity, orientation, solidity, extent, axis lengths, compactness, and bounding box analysis. The resulting features offer a high-level summary of visual shape, supporting precise detection of scratch-like patterns versus round or noisy blobs.



```python

def add_image_shape_features(df_wafer):
    df = df_wafer.copy()

    x_vals = sorted(df['DieX'].unique())
    y_vals = sorted(df['DieY'].unique())
    x_map = {x: i for i, x in enumerate(x_vals)}
    y_map = {y: i for i, y in enumerate(y_vals)}

    grid = np.zeros((len(y_vals), len(x_vals)), dtype=np.uint8)

    for row in df.itertuples():
        if not row.IsGoodDie:
            grid[y_map[row.DieY], x_map[row.DieX]] = 1

    labeled = label(grid, connectivity=2)
    props = regionprops(labeled)

    label_to_feature = {}
    for region in props:
        label_id = region.label
        label_to_feature[label_id] = {
            'eccentricity': region.eccentricity,
            'compactness': (region.perimeter ** 2) / region.area if region.area > 0 else 0,
            'major_axis': region.major_axis_length,
            'minor_axis': region.minor_axis_length,
            'orientation': region.orientation,
            'solidity': region.solidity,
            'extent': region.extent,
            'bbox_area': region.bbox_area,
            'bbox_ratio': region.bbox_area / (region.area + 1e-6)  # ratio between bbox area and actual area
        }

    features = {
        'DefectEccentricity': [],
        'DefectCompactness': [],
        'DefectMajorAxisLength': [],
        'DefectMinorAxisLength': [],
        'DefectOrientation': [],
        'DefectSolidity': [],
        'DefectExtent': [],
        'DefectBoundingBoxArea': [],
        'DefectBBoxToAreaRatio': []
    }

    for row in df.itertuples():
        x = x_map[row.DieX]
        y = y_map[row.DieY]
        lbl = labeled[y, x]

        feats = label_to_feature.get(lbl, None)
        for key in features.keys():
            features[key].append(feats.get(key, 0) if feats else 0)

    for col, values in features.items():
        df[col] = values

    return df

for name in wafer_dict:
    wafer_dict[name] = add_image_shape_features(wafer_dict[name])

print("✅ Added: 'DefectEccentricity', 'DefectCompactness', 'DefectMajorAxisLength', 'DefectOrientation'")
```

#17.BadDensity_X/Y
This extended projection-based feature suite computes per-axis densities for both bad and good dies, along with normalized values, ratios, gradients, and distance metrics from peak defect zones. The inclusion of projection rank, spread statistics, and compound scores provides deep axial insight into defect distribution, crucial for detecting linear scratch patterns and edge clustering.



```python
def add_projection_density(df_wafer, normalize=True, include_advanced=True):
    df = df_wafer.copy()

    total_x = df['DieX'].nunique()
    total_y = df['DieY'].nunique()

    bad_df = df[df['IsGoodDie'] == False]
    good_df = df[df['IsGoodDie'] == True]

    bad_by_x = bad_df['DieX'].value_counts().sort_index()
    bad_by_y = bad_df['DieY'].value_counts().sort_index()
    good_by_x = good_df['DieX'].value_counts().sort_index()
    good_by_y = good_df['DieY'].value_counts().sort_index()

    all_x = df['DieX'].unique()
    all_y = df['DieY'].unique()

    bad_by_x_full = bad_by_x.reindex(sorted(all_x), fill_value=0)
    bad_by_y_full = bad_by_y.reindex(sorted(all_y), fill_value=0)
    good_by_x_full = good_by_x.reindex(sorted(all_x), fill_value=0)
    good_by_y_full = good_by_y.reindex(sorted(all_y), fill_value=0)

    # Map counts to all dies
    df['BadDensity_X'] = df['DieX'].map(bad_by_x_full).fillna(0).astype(int)
    df['BadDensity_Y'] = df['DieY'].map(bad_by_y_full).fillna(0).astype(int)
    df['GoodDensity_X'] = df['DieX'].map(good_by_x_full).fillna(0).astype(int)
    df['GoodDensity_Y'] = df['DieY'].map(good_by_y_full).fillna(0).astype(int)

    # Ratios
    df['BadGoodRatio_X'] = df['BadDensity_X'] / (df['GoodDensity_X'] + 1)
    df['BadGoodRatio_Y'] = df['BadDensity_Y'] / (df['GoodDensity_Y'] + 1)

    # Normalization
    if normalize:
        df['BadDensity_X_Norm'] = df['BadDensity_X'] / total_y
        df['BadDensity_Y_Norm'] = df['BadDensity_Y'] / total_x
        df['GoodDensity_X_Norm'] = df['GoodDensity_X'] / total_y
        df['GoodDensity_Y_Norm'] = df['GoodDensity_Y'] / total_x

    if include_advanced:
        # Global stats
        bad_x_max = bad_by_x_full.max()
        bad_y_max = bad_by_y_full.max()
        bad_x_argmax = bad_by_x_full.idxmax()
        bad_y_argmax = bad_by_y_full.idxmax()
        bad_x_std = bad_by_x_full.std()
        bad_y_std = bad_by_y_full.std()

        # Distance from max
        df['DistanceFromMaxBadDensity_X'] = abs(df['DieX'] - bad_x_argmax)
        df['DistanceFromMaxBadDensity_Y'] = abs(df['DieY'] - bad_y_argmax)

        # Std repeated
        df['BadDensity_X_Std'] = bad_x_std
        df['BadDensity_Y_Std'] = bad_y_std

        # Gradients (difference with previous)
        bad_x_grad = bad_by_x_full.diff().fillna(0)
        bad_y_grad = bad_by_y_full.diff().fillna(0)
        df['BadDensity_X_Gradient'] = df['DieX'].map(bad_x_grad).fillna(0)
        df['BadDensity_Y_Gradient'] = df['DieY'].map(bad_y_grad).fillna(0)

        # Rank
        bad_x_rank = bad_by_x_full.rank(method='dense').astype(int)
        bad_y_rank = bad_by_y_full.rank(method='dense').astype(int)
        df['BadDensity_X_Rank'] = df['DieX'].map(bad_x_rank).fillna(0).astype(int)
        df['BadDensity_Y_Rank'] = df['DieY'].map(bad_y_rank).fillna(0).astype(int)

        # Combined projection scores
        df['BadDensity_Sum_XY'] = df['BadDensity_X'] + df['BadDensity_Y']
        df['BadGoodRatio_Sum'] = df['BadGoodRatio_X'] + df['BadGoodRatio_Y']

    return df

for name in wafer_dict:
    wafer_dict[name] = add_projection_density(wafer_dict[name])

print("✅ Added features: 'BadDensity_X' and 'BadDensity_Y'")
```

#18 add_edge_features_combined
This ultimate edge feature function merges spatial distance metrics and band-based defect density analysis. It computes raw and normalized distances from each die to wafer edges, flags dies near the border, supports multi-width ring-based encoding, and summarizes bad/good ratios in the edge zone. These features are essential for differentiating central from peripheral defects and refining edge-aware decision logic.




```python
def add_edge_features_combined(df_wafer, edge_threshold=2, normalize=True, include_density=True, multi_band=True):
    df = df_wafer.copy()

    min_x, max_x = df['DieX'].min(), df['DieX'].max()
    min_y, max_y = df['DieY'].min(), df['DieY'].max()
    wafer_width = max_x - min_x + 1
    wafer_height = max_y - min_y + 1

    # --- Distance Features ---
    df['DistanceToXEdge'] = df['DieX'].apply(lambda x: min(abs(x - min_x), abs(x - max_x)))
    df['DistanceToYEdge'] = df['DieY'].apply(lambda y: min(abs(y - min_y), abs(y - max_y)))
    df['DistanceToAnyEdge'] = df[['DistanceToXEdge', 'DistanceToYEdge']].min(axis=1)

    # --- Normalized Distances ---
    if normalize:
        df['DistanceToXEdge_Norm'] = df['DistanceToXEdge'] / (wafer_width / 2)
        df['DistanceToYEdge_Norm'] = df['DistanceToYEdge'] / (wafer_height / 2)
        df['DistanceToAnyEdge_Norm'] = df['DistanceToAnyEdge'] / (min(wafer_width, wafer_height) / 2)

    # --- Binary Edge Proximity Indicator ---
    df['IsNearEdge'] = (df['DistanceToAnyEdge'] <= edge_threshold).astype(int)

    # --- Multi-band Edge Zones ---
    if multi_band:
        for r in range(1, edge_threshold + 1):
            df[f'IsInEdgeBand_r{r}'] = (df['DistanceToAnyEdge'] == r).astype(int)

    # --- Edge Density Analysis ---
    if include_density:
        # Edge zone indicator
        df['IsInEdgeBand'] = (df['DistanceToAnyEdge'] <= edge_threshold).astype(int)
        edge_df = df[df['IsInEdgeBand'] == 1]

        total_edge_dies = len(edge_df)
        bad_edge_dies = (edge_df['IsGoodDie'] == False).sum()
        good_edge_dies = (edge_df['IsGoodDie'] == True).sum()

        bad_density = bad_edge_dies / total_edge_dies if total_edge_dies > 0 else 0
        good_density = good_edge_dies / total_edge_dies if total_edge_dies > 0 else 0
        bad_good_ratio = bad_edge_dies / (good_edge_dies + 1)

        # Assign global density to all dies
        df['EdgeBand_BadDensity'] = bad_density
        df['EdgeBand_GoodDensity'] = good_density
        df['EdgeBand_BadGoodRatio'] = bad_good_ratio

        # Local binary indicator: bad and in edge
        df['IsBadInEdgeBand'] = ((df['IsInEdgeBand'] == 1) & (df['IsGoodDie'] == False)).astype(int)

    return df


for name in wafer_dict:
    wafer_dict[name] = add_edge_features_combined(
        wafer_dict[name],
        edge_threshold=2,
        normalize=True,
        include_density=True,
        multi_band=True         #
    )

print("✅ Added full edge-based features to all wafers.")

```

#19 LocalEntropy
This fast version of the local entropy extractor uses convolution-style filtering to compute binary entropy, mean bad die ratio, and standard deviation in a 3×3 neighborhood for each die. These features characterize the local structure around each die, highlighting uncertainty, density, and inconsistency in spatial patterns.



```python


def compute_entropy_block(block):
    """Compute binary entropy of 3x3 block (flattened to length 9)."""
    counts = np.bincount(block.astype(int), minlength=2)
    probs = counts / counts.sum() if counts.sum() > 0 else np.array([0.5, 0.5])
    return scipy_entropy(probs, base=2)

def add_local_entropy_fast(df_wafer):
    df = df_wafer.copy()

    # Create matrix for bad=1, good=0
    x_vals = sorted(df['DieX'].unique())
    y_vals = sorted(df['DieY'].unique())
    x_map = {x: i for i, x in enumerate(x_vals)}
    y_map = {y: i for i, y in enumerate(y_vals)}

    matrix = np.zeros((len(y_vals), len(x_vals)), dtype=np.uint8)
    for row in df.itertuples():
        matrix[y_map[row.DieY], x_map[row.DieX]] = int(not row.IsGoodDie)

    # Apply entropy filter over 3x3 window
    entropy_matrix = generic_filter(matrix, compute_entropy_block, size=3, mode='constant', cval=0)

    # Additional features: local mean and std (texture indicators)
    local_mean = generic_filter(matrix.astype(float), np.mean, size=3, mode='constant', cval=0)
    local_std = generic_filter(matrix.astype(float), np.std, size=3, mode='constant', cval=0)

    # Map back to DataFrame
    entropy_list = []
    mean_list = []
    std_list = []

    for row in df.itertuples():
        xi = x_map[row.DieX]
        yi = y_map[row.DieY]
        entropy_list.append(entropy_matrix[yi, xi])
        mean_list.append(local_mean[yi, xi])
        std_list.append(local_std[yi, xi])

    df['LocalEntropy'] = entropy_list
    df['LocalBadRatio_3x3'] = mean_list
    df['LocalBadStd_3x3'] = std_list

    return df

for name in wafer_dict:
    wafer_dict[name] = add_local_entropy_fast(wafer_dict[name])

print("✅ Added feature: 'LocalEntropy'")
```

```python
merged_df['keep'] = (merged_df['CombinedScore'] > 0.01).astype(int)

X_logreg = merged_df[['Importance', 'MeanAbsSHAP', 'CombinedScore']]
y_logreg = merged_df['keep']

model = LogisticRegression()
model.fit(X_logreg, y_logreg)

merged_df['keep_proba'] = model.predict_proba(X_logreg)[:, 1]

fpr, tpr, thresholds = roc_curve(y_logreg, merged_df['keep_proba'])
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("🔍 ROC Curve for Feature Keep Decision")
plt.legend(loc="lower right")
plt.grid(True)
plt.tight_layout()
plt.show()


j_scores = tpr - fpr
best_idx = j_scores.argmax()
best_threshold = thresholds[best_idx]
print(f"🎯 Best threshold by Youden's J: {best_threshold:.4f}")


merged_df['keep_final'] = (merged_df['keep_proba'] >= best_threshold).astype(int)
selected_features_17= merged_df[merged_df['keep_final'] == 1]['Feature'].tolist()
print(f"✅ Selected {len(selected_features_16)} features using threshold {best_threshold:.4f}")
```
